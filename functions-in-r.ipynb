# Funtions in R

#### A function is a block of code used to execute a specified task, arrange in an organized format.
#### Functions can be reused whenever required in a program.
#### R has both an array & built-in functions for ready use by user & user defined functions.

* ### Basic Mathematical Built-in functions 

```r
# Absolute value 
a<-20
abs(20)
```

```r
# Log values of a to the base b.
# If base is not specified, then natural logarithm is returned as the result.
b<-2
log(a,base=b)
```

```r
# Exponential value
exp(a)
```

```r
# Square root
sqrt(a)
```

```r
# Factorial (b!)
factorial(b)
```

* ### Basic Statistical Built-in functions

```r
# Mean
x<-c(3,2,5,6,4,8)
mean(x)
```

```r
# Median
median(x)
```

```r
# Variance
var(x)
```

```r
#Standard deviation
sd(x)
```

```r
# Divides dataset into quantile
quantile(x)
```

* ### General Purpose Built-in Functions

```r
# Minimum and Maximum values
range(x)
```

```r
# Sorting data in ascending order
sort(x)
```

```r
# Return the order of value in dataset
order(x)
```

```r
# Length of dataset
length(x)
```

```r
# sign of data
s<- -20 
sign(s)
```

```r
# Convert to lowercase
u <- 'ABCD'
tolower(u)
```

```r
# Convert to uppercase
l <- 'abcd'
toupper(l)
```

```r
# Create console screen
ls()
```

* ### User-Defined Functions
You can create your own function for certain application. Like

#### function_name <- function(argument list){

#### body of function

#### }


```r
# Lets create a simple addition function
do_addition<- function(o,p){
    answer <- o+p
    print(paste("The answer is", answer))
}
```

```r
do_addition(20,30)
```
