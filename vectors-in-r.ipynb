# Vectors in R

### Vector is a basic data stucture in R. It is a sequence of elements.
### Elements in a vector can be heterogeneous(Vector can have elements with different datatype).

* **Atomic vector** is one that has all the elements of the **same datatype**(Homogeneous).
* Vector with Heterogeneous datatype is known as **List**. 

#### **Vector creation methods**

#### 1.**c() function**
It combines the elements and returns a vector in the form of a 1D array.

```r
x<- c(10,20,30)
x
```

#### 2. **':' operator** 
This operator assigns the numbers in given range

```r
y<-10:20
y
```

#### 3. **seq() function**
Assigns numbers in a given range with an increment.

```r
z<- seq(20,30, by=2)
z
```

#### **Vector operation**

```r
# Accessing an element
q <- c(12,34,78,53,567,98)
q[3] # Accesing the 3rd element
```

```r
# Accessing a series of elements
q[1:5]
```

```r
# Negative index
q[-2] # Displays all elements except 2nd 
```

```r
# Replacing a values in a vector
q[5]<-23 # Replaces 5th value to 23
q
```

```r
# Replacing values with an expression
q[q<35]<-10 #replaces the elements less than 35 to 10
q
```

```r
# Truncating values in vector
q<-q[2:6] #Deleting 1st value
q
```

```r
# Adding two vectors
p<-c(10,10,10,10,10)
r<-p+q
r
```

```r
# Adding a values to vector
s<-p+2
s
```

```r
# Deleting vector
q<-NULL
q
```
