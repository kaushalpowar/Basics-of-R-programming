# Data types in R

#### Data structure form the basic for any programming language. Therefore it is essential to understand Data structures befor diving into programming language. 
#### R also provides function ***class()*** to check the data type of particular variable.




#### There are five different data type in R.
1. Numeric
2. Integer
3. Complex
4. Character 
5. Logical

### 1. Numeric
Numeric data type store decimal values.
The ***as.numeric()*** function is used to convert any data type into Numeric.

```r
x <- 12.3
class(x)
```

```r
y <- as.numeric(10)
y
```

```r
class(y)
```

### 2. Integer
Integer data type assigns variable without decimal digits.

Assigning the alphabet "L" at end of number will make it integer.

The ***as.integer()*** function used to convert any data type into integer.

```r
#Assigning L at the eng to make it integer
a <- 20L
a
class(a)
```

```r
b <- as.integer(21.1)
b
class(b)
```

### 3. Complex

Complex data type contain two numeric values with an imaginary nuber "i".

```r
c <- 3+4i
class(c)
```

### 4. Character
Character data type is used to represent strings.

Strings should be written eithin single or double quotes.

The ***as.string()*** function will convert any data type into character.

```r
d <- "This is character"
class(d)
```

```r
e <- 2022
f <- as.character(e)
f
class(f)
```

### 5. Logical
Logical data type stores two values: TRUE or FALSE.

```r
g <- TRUE
class(g)
```
