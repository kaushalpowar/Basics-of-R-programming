# Operators in R

#### Operators indicate the mathematical operations or logical manipulations that can be performed.
#### The R software is bulit-in with different operators that help user to build numeric & logical expressions.

#### There are four built-in operators R.
1. Assignment operators
2. Arithmetic operators
3. Relational operators
4. Logical operatios


### 1. Assignment Operators
Assignment operators are used to assign values to the variables.
There 3 types of Assignment operators.

* First type

```r
## Assignment rightwards
x <- 10 
```

```r
## Assignment leftwards
20 -> y
```

* Second type

```r
# Assignment
a = 30
```

* Third type

```r
## Assignment rightwards
v <<-40
```

```r
# Assignment leftwards
50 ->> z
```

### 2. Arithmetic Operators
Arithmetic operators perform simple mathematical operations like 
* Addition
* Subtraction
* Multiplication
* Division
* Exponent
* Modulus operator
* Integer division

```r
# Addition
a<-10
b<-20
c<-a+b
c
```

```r
# Subtraction
a-b
```

```r
# Multiplication
a*b
```

```r
# Division
a/b
```

```r
# Exponent
a**2
#a^2
```

```r
# Modulus operator
a%%b
```

```r
# Integer division
b%/%a
```

### 3. Relational Operators
Relational operators compares two variables or values & gives result as boolean value(TRUE/FALSE).

```r
# Less than
a>b
```

```r
# Greater than
a<b
```

```r
# Less than or equal to
a<=b
```

```r
# Greater than or equal to
a>=b
```

```r
# Equal to
a==b
```

```r
# Not equal to
a!=b
```

### 4. Logical Operators 
Logical operators perform operation like AND, OR, NOT. They provide Boolean value as output.

```r
# Logical NOT
x<-c(TRUE, TRUE)
y<-c(FALSE,FALSE)
!x
```

```r
# Element-wise logical AND
x&y 
# It returns FALSE FALSE cause both the element in present in list x & list y are not same.
```

```r
# Logical AND(compares the first element in vector)
x&&y
```

```r
# Element-wise logical OR
x|y
```

```r
# Logical OR (compares the first element in vector)
x||y
```
