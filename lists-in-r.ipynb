# Lists in R

#### A list contains elements with different data types such as numbers, strings, characters, or vectors within it.
#### ***list()*** fucntion is used to create list.

```r
# Creating a list
list1 <- list("A","B",c(10,20,30), c(TRUE, FALSE),123.4,56.7)
```

```r
# Accessing first element from list
list1[1]
```

```r
# Accessing third element from list
list1[3]
```

```r
# Accessing first 3 elements
list1[1:3]
```
