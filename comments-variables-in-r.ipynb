# Comments & Variables in R

## Comments in R

#### Comment statments are types with # in the begining(just like Python).
#### Any statement starting with # symbol will be ignored & will not be executed.
#### It is a good practice to use # for **documentation** & **Explanation** of code.

```r
# This is comment & it will be ignored while executing.
a <- 10
b <-20
c <- a+b
c
```
